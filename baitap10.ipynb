```python
#W10A1
def divide(a, b) :
    try :
        result = a/b
        return f"{result:.2f}"
    except ZeroDivisionError :
        return "Không thể chia cho 0"
x, y = map(int,input().split())
print(divide(x, y))
```

```python
#W10A3
def Is_happy_num() :
    try :
        n = int(input("Nhập số nguyên dương a :"))
        checked = set()
        while n != 1 and n not in checked :
            checked.add(n)
            total = 0
            while n > 0 :
                fac = n%10
                total += fac**2
                n //= 10
            n = total
        if n == 1 :
            print("YES")
        else : 
            print("NO")
    except ValueError :
        print("Dữ liệu nhập vào không phải số nguyên")
if __name__ == "__main__" :
    Is_happy_num()
    pass
```

```python
#W10A5
def list_to_int(a) :
    b = []
    for i in a :
        try :
            num = int(i)
            b.append(num)
        except (ValueError, TypeError) :
            print(f"Can't convert '{i}' to int. Skipped")
    return b
n = eval(input())
print(list_to_int(n))
```

```python
#W10A6
from math import sqrt
def square_root(n) :
    try :
        result = sqrt(n)
        return f"{result:.2f}"
    except ValueError :
        return "Số âm"
a = int(input())
print(square_root(a))
```

```python
#W10A7
def yrs_old(n) :
    try :
        age = int(n)
        if age < 0 :
            return "Tuổi không hợp lệ"
        else :
            BOD = 2025 - age
            return BOD
    except ValueError :
        return "Tuổi không hợp lệ"
a = input()
print(yrs_old(a))
```

```python
#W10A8
def ktraduoi(n):
  duoi = ""
  for i in range(len(n)):
    if n[i] == ".":
      for j in range(i+1,len(n),1):
        duoi = duoi + n[j]
  if duoi[0] in "tT" and duoi[1] in "xX" and duoi[2] in "tT":
    return "YES"
  if duoi[0] in "zZ" and duoi[1] in "iI" and duoi[2] in "pP":
    return "YES"
  else:
    return "No"
n = list(input())
try:
  if ktraduoi(n) == "No":
    raise ValueError
  print("Doc file thanh cong")
except (ValueError,IndexError):
  print("File khong hop le")
```

```python
#W10A11
def is_valid_regex(s):
    repeat = {'*', '+', '?'}
    i = 0
    n = len(s)
    while i < n:
        if s[i] == '\':
            i += 2
            continue
        if s[i] in repeat:
            if i == 0:
                return False
            if s[i-1] in repeat:
                return False
        i += 1
    return True
T = int(input())
for _ in range(T):
    S = input().strip()
    print("True" if is_valid_regex(S) else "False")
```
