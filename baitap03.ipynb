```python
#Largest number
a, b, c, d = map(int, input("Nhập vào 4 số nguyên a, b, c và d : ").split())
max = a
if b > a :
    max = b
if c > a : 
    max = c
if d > a :
    max = d
print(max)
```

```python
#Sorting triangle
a, b ,c = map(int, input("Nhập ba số nguyên dương a, b và c : ").split())
if a + b <= c or a + c <= b or b + c <= a :
    print("Không phải tam giác")
else :
    if a == b == c :
        print("Tam giác đều")
    elif a == b != c or b == c != a or a == c != b :
        print("Tam giác cân")
    else : 
        print("Tam giác thường")
```

```python
#Electricity price calculation
n = int(input("Nhập số n kWh điện tiêu thụ : "))
if 0 <= n <= 50 :
    price = 1500*n
if 51 <= n <=100 :
    price = 1500*50 + 2000*(n-50)
if n > 100 :
    price = 1500*50 + 2000*50 + 3000*(n-100)
print(price)
```

```python
#Linear equation solving
a = float(input("Nhập số thực a : "))
b = float(input("Nhập số thực b : "))
if a == 0 :
    if b != 0 :
        print("VÔ nghiệm")
    else :
        print("Vô số nghiệm")
else : 
    x = -b/a
    print(f"{x:.2f}")
```
