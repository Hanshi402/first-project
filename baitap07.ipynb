```python
#W7A1
def binary_search(arr, value):
    left, right = 0, len(arr) - 1
    while left <= right: 
        mid = (left + right) // 2
        if arr[mid] == value:
            return mid 
        elif arr[mid] < value:
            left = mid + 1 
        else:
            right = mid - 1 
    return -1 
a = list(map(int,input().split()))
n = int(input())
stat = binary_search(a, n)
print(stat)
```

```python
#W7A2
def count_occurrences(arr, x) -> int :
    count = 0
    for i in range(len(arr)) :
        if arr[i] == x :
            count += 1
    return count
a = list(map(int, input().split()))
n = int(input())
print(count_occurrences(a, n))
```

```python
#W7A3
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr
a = list(map(int, input().split()))
print(bubble_sort(a))
```

```python
#W7A8
def checkdieukien(n,ds):
  if ds[0]<=n<=ds[1]:
    return ds[1] - ds[0] + 1
  return "No"
def minof(ds):
  min = ds[0]
  for i in range(1,len(ds)):
    if ds[i] <= min:
      min = ds[i]
  return min
interval = eval(input())
queries = eval(input())
dscuoi = []
for i in queries:
  ds = []
  for j in interval:
    if checkdieukien(i,j) != "No":
      ds.append(checkdieukien(i,j))
  if len(ds) == 0:
    dscuoi.append(-1)
  else:
    dscuoi.append(minof(ds))
print(dscuoi)
```

```python
#W7A9
class Node():
    def __init__(self, key, pos, left):
        self.left = None
        self.right = None
        self.val = key
        self.pos = pos 
        self.rot = left
count = 0
class BST:
    def __init__(self):
        self.root = None
        self.last = 0
    def insert(self, root, key):
        global count
        if root is None:
            node = Node(key, count, 0)
            self.last = node.pos
            count = 0
            return node
        else:
            root.pos = max(count+root.rot, root.pos)
            if root.val < key:
                count = root.pos + 1
                root.right = self.insert(root.right, key)
            else:
                root.pos += 1
                root.rot += 1
                root.left = self.insert(root.left, key)
        return root
a=list(map(int,input().split()))
a=a[::-1]
tree=BST()
root = None
res=[]
for i in a:
    root=tree.insert(root, i)
    res.append(tree.last)
res=res[::-1]
print(res)
```

```python
#W7A10
def maxof(ds):
  max = ds[0]
  for i in range(1,len(ds)):
    if ds[i] >= max:
      max = ds[i]
  return max
nums = eval(input())
n = int(input())
ds = []
for i in range(len(nums)):
  for k in range(len(nums)-1,i-1,-1):
     ds.append(sum(nums[i:k])%n)
print(maxof(ds))
```

```python
#W7A13
def demphantu(dsach,ptu):
  count = 0
  for i in dsach:
    if i == ptu:
      count = count + 1
  return count
def maxof(ds):
  max = ds[0]
  for i in range(1,len(ds)):
    if ds[i] >= max:
      max = ds[i]
  return max
ds1 = list(map(int,input().split()))
ds2 = list(set(ds1))
dscap = []
for i in ds2:
  cap = []
  k = i + 1
  if k in ds2:
    cap.append(i)
    cap.append(k)
    dscap.append(cap)
dsdodai = []
for i in dscap:
  dodai = demphantu(ds1,i[0]) + demphantu(ds1,i[1])
  dsdodai.append(dodai)
print(maxof(dsdodai))
```
