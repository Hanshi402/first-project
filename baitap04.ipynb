```python
#W4A7
from math import isqrt
def is_prime(n) :
    if n < 2 :
        return False
    for i in range(2, isqrt(n) + 1) :
        if n%i == 0 :
            return False
    return True
n = int(input("Nhập số nguyên dương n : "))
while n < 2 :
    n = int(input("Nhập số nguyên dương n : "))
Factor = []
for i in range (1, n + 1) :
    if n%i == 0 :
        Factor.append(i)
Prime_factor = []
for n in Factor :
    if is_prime(n) :
        Prime_factor.append(n)
print(Prime_factor[::-1][0])
#Max of list :
def max_of_list() :
    max_factor = Prime_factor[0]
    for x in range(len(Prime_factor)-1) :
        if Prime_factor[x+1] > Prime_factor[x] :
            max_factor = Prime_factor[x+1]
    print(max_factor)
```

```python
#W4A9
def check(x):
    for j in range(len(x)):
        for k in range(j + 1, len(x)):
            if x[j] == x[k]:
                return False
    return True
n = int(input())
List_print = []

for i in range(1, n+1):
    if i ** 0.5 == int(i ** 0.5):
        if check(list(str(i))):
            List_print.append(i)
if len(List_print) == 0:
    print("no number")
else:
    print(*List_print)
```

```python
#W4A10
def collatz(n):
    lenght = [n]
    while n != 1:
        n = n//2 if n%2==0 else 3*n+1
        lenght.append(n)
    return lenght

n = int(input())
find_max = []
find_x_min = []

for y in range(1, n+1):
    find_max.append(len(collatz(y)))
a = max(find_max)

for x in range(1, n+1):
    if len(collatz(x)) == a:
        find_x_min.append(x)

print(min(find_x_min))
print(a)
```

```python
#W4A26
def fibo(x):
    fibo_list = []
    for i in range(x):
        if i == 0 or i == 1:
            fibo_list.append(1)
        else:
            fibo_list.append(int(fibo_list[i-1]) + int(fibo_list[i-2]))
    return fibo_list   

A = int(input())
x = 1

while max(fibo(x)) <= A:
    x += 1
print(max(fibo(x-1)))
```
