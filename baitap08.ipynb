```python
#W8A1
class Cylinder:
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height
        self.pi = 3.14
    def surface_area(self):
        return 2 * self.pi * self.radius**2 + 2 * self.pi * self.radius * self.height
    def volume(self):
        return self.pi * self.radius**2 * self.height
r, h = map(float, input().split())
cyl = Cylinder(r, h)
print(f"{cyl.surface_area():.2f} {cyl.volume():.2f}")
```

```python
#W8A2
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year

    def is_leap_year(self):
        return (self.year % 400 == 0) or (self.year % 4 == 0 and self.year % 100 != 0)

    def days_in_month(self):
        if self.month in [1, 3, 5, 7, 8, 10, 12]:
            return 31
        elif self.month in [4, 6, 9, 11]:
            return 30
        elif self.month == 2:
            return 29 if self.is_leap_year() else 28
        else:
            return 0
        
    def is_valid(self):
        if self.year < 1:
            return False
        if self.month < 1 or self.month > 12:
            return False
        if self.day < 1 or self.day > self.days_in_month():
            return False
        return True

    def next_day(self):
        self.day += 1
        if self.day > self.days_in_month():
            self.day = 1
            self.month += 1
            if self.month > 12:
                self.month = 1
                self.year += 1

    def __str__(self):
        return f"{self.day:02d}/{self.month:02d}/{self.year}"

try:
    input_str = input("Enter date (dd/mm/yyyy): ").strip()
    parts = input_str.split('/')
    if len(parts) != 3:
        raise ValueError 
    d, m, y = map(int, parts)
    date_obj = Date(d, m, y)
    if date_obj.is_valid():
        date_obj.next_day()
        print(date_obj)
    else:
        print("INVALID")
except ValueError:
    print("INVALID")
```

```python
#W8A3
from math import sqrt
class pointR2:
  def __init__(self,x,y):
    self.x=x
    self.y=y
  def SpecialPointCheck(self):
    if self.x == 0 and self.y == 0:
      return "Đây là gốc tọa độ"
    if self.x == 0:
      return "Nằm trên Oy"
    if self.y == 0:
      return "Nằm trên Ox"
    else:
      return ""
  def DistanceToTheOrigin(self):
    a = self.x
    b = self.y
    kc = sqrt(a**2+b**2)
    return f"khoảng cách tới O là {round(kc,2)}"
A = pointR2(7,8)
print(A.SpecialPointCheck())
print(A.DistanceToTheOrigin())
```

```python
#W8A5
class Shoppingcart:
  def __init__ (self):
    self.sp=[]
  def them_sp(self,ten,gia):
    self.sp.append({"ten":ten,"gia":gia})
    print(f"đã thêm {ten} với giá {gia} vào giỏ hàng")
  def xoa_sp(self,ten):
    for sp in self.sp:
      if sp["ten"]==ten:
        self.sp.remove(sp)
        print(f"{ten} đã được xóa khỏi giỏ hàng")
  def kt_rong(self):
    if len(self.sp)==0:
      print("giỏ hàng hiện đang rỗng")
    else:
      print("giỏ hàng hiện không rỗng")
  def tien_tong(self):
    tong=0
    for sp in self.sp:
      tong+=sp["gia"]
    print(f"tổng tiền là {tong}")
  def hien_thi(self):
    print("danh sách sản phẩm trong giỏ hàng:")
    for sp in self.sp:
      print(f"{sp["ten"]} với giá {sp["gia"]}")
  def xoa(self):
    self.sp.clear()
    print("giỏ hàng đã được xóa")
  def chuong_trinh_chinh(self):
    Sh=Shoppingcart()
    while True:
      print("1.Thêm sản phẩm")
      print("2.Xóa sản phẩm")
      print("3.Kiểm tra giỏ hàng rỗng")
      print("4.Tính tiền tổng")
      print("5.Hiển thị giỏ hàng")
      print("6.Xóa giỏ hàng")
      n=int(input("chọn chức năng: "))
      if n==1:
        ten=input("nhập tên sản phẩm: ")
        gia=float(input("nhập giá sản phẩm: "))
        Sh.them_sp(ten,gia)
      elif n==2:
        ten=input("nhập tên sản phẩm:")
        Sh.xoa_sp(ten)
      elif n==3:
        Sh.kt_rong()
      elif n==4:
        Sh.tien_tong()
      elif n==5:
        Sh.hien_thi()
      elif n==6:
        Sh.xoa()
      thoat=(input("bạn có muốn thoát không?"))
      if thoat.lower()=="có":
        break
if __name__=="__main__":
    Sh=Shoppingcart()
    Sh.chuong_trinh_chinh()
```

```python
#W8A6
class Rectangle :
    def __init__(self, w, h) :
        self.w = w
        self.h = h
    def area(self) : 
        return self.w * self.h
    def perimeter(self) :
        return (self.w + self.h) * 2
    def scale(self, k) :
        self.w = self.w * k
        self.h = self.h * k
values = input()
w, h, k = map(float, values.split())
new_rectangle = Rectangle(w, h)
new_rectangle.scale(k)
print(f"{new_rectangle.area():.2f} {new_rectangle.perimeter():.2f}")
```

```python
#W8A8
a,b = map(str,input("Nhập tên người sở hữu và số dư : ").split())
n = int(input("Nhập số lệnh : "))
class BankAccount:
  def __init__(self,owner,balance):
    self.owner = owner
    self.balance = balance
  def deposit(self,amount):
      self.balance = self.balance + amount
  def withdraw(self,amount):
    if amount > self.balance:
      self.balance = self.balance
    else:
      self.balance = self.balance - amount
  def __str__(self):
    return f"{self.balance:.2f}"
acc1 = BankAccount(a,float(b))
count = 1
while count <= n:
  lenh = input("Nhập lệnh : ")
  hanhdong,soluong = map(str,lenh.split())
  if hanhdong == "DEPOSIT" :
    acc1.deposit(float(soluong))
  if hanhdong == "WITHDRAW" :
    acc1.withdraw(float(soluong))
  count = count + 1
print(acc1.__str__())
```

```python
#W8A9
name = input("Nhập tên học sinh : ")
n = int(input("Số môn nhập điểm : "))
class Student:
  def __init__(self,name):
    self.name = name
    self.dic = {}
  def add_score(self,subject,score):
    self.dic[subject] = score
  def avg(self):
    tong = 0
    for v in self.dic.values():
      tong = tong + v
    tb = tong/n
    return round(tb,2)
  def rank(self,tb):
    if tb >= 8:
      return "Excellent"
    if 6.5 <= tb < 8:
      return "Good"
    if 5 <= tb < 6.5:
      return "Average"
    else:
      return "Poor"
s1 = Student(name)
count = 1
while count <= n:
  mon,diem = map(str,input(f"Nhập dữ liệu môn {count} : ").split())
  s1.add_score(mon,float(diem))
  count = count + 1
print(f"{name} {s1.avg()} {s1.rank(s1.avg())}")
```

```python
#W8A10
class Book:
  def __init__(self,title,author,year):
    self.title = title
    self.author = author
    self.year = year
    self.sach = []
class Library(Book):
  def add(self):
    them = []
    them.append(self.title)
    them.append(self.author)
    them.append(self.year)
    self.sach.append(them)
  def count_by_author(self,name):
    count = 0
    for i in self.sach:
      if i[1] == name:
        count = count + 1
    return count
  def find_by_year(self,y):
    count = 0
    for i in self.sach:
      if i[2] == y:
        count = count + 1
      return count
n = int(input("Nhập số thao tác : "))
count = 1
while count <= n:
  a1,a2 = map(str,input("Nhập lệnh : ").split())
  if a1 == "ADD":
    a,b,c = map(str,a2.split(";"))
    Library(a,b,c).add()
  if a1 == "COUNT":
    print(Library(a,b,c).count_by_author(a2))
  if a1 == "COUNTYEAR":
    print(Library(a,b,c).find_by_year(a2))
  count = count + 1
```

```python
#W8A11
import numpy as np
xA,yA,zA = map(float,input("A:").split())
xB,yB,zB = map(float,input("B:").split())
xC,yC,zC = map(float,input("C:").split())
xD,yD,zD = map(float,input("D:").split())
class Point:
  def __init__(self,x,y,z):
    self.x=x
    self.y=y
    self.z=z
  def NormalVectorOfPlane(self,other1,other2):
    v1 = np.array([self.x-other1.x,self.y-other1.y,self.z-other1.z])
    v2 = np.array([self.x-other2.x,self.y-other2.y,self.z-other2.z])
    return np.cross(v1,v2)
  def AngleBetween2Planes(v1,v2):
    V1 = np.linalg.norm(v1)
    V2 = np.linalg.norm(v2)
    angle = np.arccos((np.dot(v1,v2))/(V1*V2))
    return f"{(angle*180/np.pi):.2f}"
A = Point(xA,yA,zA)
B = Point(xB,yB,zB)
C = Point(xC,yC,zC)
D = Point(xD,yD,zD)
n1 = A.NormalVectorOfPlane(B,C)
n2 = B.NormalVectorOfPlane(C,D)
print(Point.AngleBetween2Planes(n1,n2))
```

```python
#W8A12
r1,i1 = map(int,input().split())
r2,i2 = map(int,input().split())
class Complex:
  def __init__(self,R=0.0,I=0.0):
    self.R = int(R)
    self.I = int(I)
  def Product(self,other):
    Real = self.R*other.R - self.I*other.I
    Imag = self.R*other.I + self.I*other.R
    return f"{Real} {Imag}"
C1 = Complex(r1,i1)
C2 = Complex(r2,i2)
print(C1.Product(C2))
```
