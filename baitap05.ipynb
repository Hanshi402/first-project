```python
#W5A3
from math import isqrt
def is_prime(n) :
    if n < 2 :
        return False
    for i in range(2, isqrt(n) + 1) :
        if n%i == 0 :
            return False
    return True
n = int(input("Nhập một số n :"))
print(is_prime(n))
```

```python
#W5A4
n = int(input("Nhập số nguyên n : "))
def is_perfect(n) :
    Factor = []
    for i in range(1, n) :
        if n % i == 0 :
            Factor.append(i)
    if n == sum(Factor) :
        return True
    else : 
        return False
print(is_perfect(n))
```

```python
#W5A4
n = int(input("Nhập số nguyên n : "))
def is_perfect(n) :
    Factor = []
    for i in range(1, n) :
        if n % i == 0 :
            Factor.append(i)
    summ = 0
    for x in Factor :
        summ += x
    if summ == n :
        return True
    else : 
        return False
print(is_perfect(n))
```

```python
#W5A6
def fac(n) :
    if n == 0 or n == 1 :
        return 1
    else :
        factorial = 1
        for i in range(2, n + 1) :
            factorial *= i
        return factorial
n = int(input())
print(fac(n))
```

```python
#W5A7
def operators(num1, num2, operat) :
    if operat == "+" :
        result = num1 + num2 
    elif operat == "-" :
        result = num1 - num2
    elif operat == "*" :
        result = num1 * num2
    elif operat == "/" :
        result = num1 / num2
    else :
        return "SyntaxERROR"
    return round(result, 2)
x, y = map(float,input().split())
z = input()
print(operators(x, y ,z))
```

```python
#W5A8
def Hamming_distance(a, b) :
    n = a^b
    count = 0
    while n > 0 :
        count += n&1
        n >>= 1
    return count
x, y = map(int,input().split())
print(Hamming_distance(x, y))
```

```python
#W5A9
def sum_of_digits(a) :
    total = 0
    while a > 0 :
        total += a%10
        a //= 10
    return total
n = int(input())
print(sum_of_digits(n))
```
